```python
import pandas as pd

# Assuming df is your dataframe
unique_files = df['FILE'].unique()

# This will hold the files that have all three labels
files_with_all_labels = []

for file in unique_files:
    # Get the subset of the dataframe for the current file
    subset_df = df[df['FILE'] == file]

    # Check if all three labels are present
    if 'SmellQuality' in subset_df['TAGS'].values and 'SmellWord' in subset_df['TAGS'].values and 'SmellSource' in subset_df['TAGS'].values:
        files_with_all_labels.append(file)

print(files_with_all_labels)
```

```python
# Create new DataFrame with split TAGS
new_columns = df['TAGS'].str.split('|', expand=True)

# Rename the columns of new DataFrame
new_columns.columns = ['tags{}'.format(i) for i in range(1, len(new_columns.columns)+1)]

# Concatenate the new DataFrame to the original DataFrame
df = pd.concat([df, new_columns], axis=1)
```

```python
df = df[['QUOTE_TRANSCRIPTION', 'tags1', 'tags2']]
```

```python
from sklearn.metrics import cohen_kappa_score

# Convert NaN values to 'None'
df['tags1'] = df['tags1'].fillna('None')
df['tags2'] = df['tags2'].fillna('None')

# Compute Cohen's Kappa
kappa = cohen_kappa_score(df['tags1'], df['tags2'])

print("Cohen's Kappa: ", kappa)
```

```python
from sklearn.metrics import cohen_kappa_score

# Identify all unique tags
all_tags = pd.concat([df['tags1'], df['tags2']]).unique()

for tag in all_tags:
    # Generate binary annotation vectors
    tags1_binary = (df['tags1'] == tag).astype(int)
    tags2_binary = (df['tags2'] == tag).astype(int)

    # Compute and print Cohen's Kappa for the current tag
    kappa = cohen_kappa_score(tags1_binary, tags2_binary)
    print(f"Cohen's Kappa for tag '{tag}': ", kappa)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Identify all unique tags
all_tags = pd.concat([df['tags1'], df['tags2']]).unique()
kappas = []

for tag in all_tags:
    # Generate binary annotation vectors
    tags1_binary = (df['tags1'] == tag).astype(int)
    tags2_binary = (df['tags2'] == tag).astype(int)

    # Compute and store Cohen's Kappa for the current tag
    kappa = cohen_kappa_score(tags1_binary, tags2_binary)
    kappas.append(kappa * 100)

# Plotting
fig, ax = plt.subplots()

# Create a color palette
colors = plt.cm.viridis(np.linspace(0, 1, len(all_tags)))

# Create a bar plot
ax.bar(all_tags, kappas, color=colors)

# Set labels
ax.set_ylabel('Cohen\'s Kappa Score (%)')
ax.set_title('Cohen\'s Kappa for each tag')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility

# Adjust y-axis to show percentages
y_values = np.arange(0, 105, 5)
ax.set_yticks(y_values)
ax.set_yticklabels(['{}%'.format(i) for i in y_values])

# Show the plot
plt.show()
```
